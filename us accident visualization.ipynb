```python
import pandas as pd

# Replace 'your_file_path' with the actual path to your CSV file
file_path = r'C:\Users\NHI652\Downloads\archive (2)\US_Accidents_March23.csv'
df = pd.read_csv(file_path)

# Display all column names in the dataset
print(df.columns)

# Step 3: Analyze Patterns

# Check if 'Road_Conditions' is present in the list of columns
if 'Road_Conditions' in df.columns:
    # Group by road conditions and analyze accident counts
    road_condition_counts = df['Road_Conditions'].value_counts()
    print(road_condition_counts)
else:
    print("Column 'Road_Conditions' not found in the dataset.")

# Check if 'Weather_Condition' is present in the list of columns
if 'Weather_Condition' in df.columns:
    # Group by weather conditions and analyze accident counts
    weather_condition_counts = df['Weather_Condition'].value_counts()
    print(weather_condition_counts)
else:
    print("Column 'Weather_Condition' not found in the dataset.")

# Check if 'Sunrise_Sunset' is present in the list of columns
if 'Sunrise_Sunset' in df.columns:
    # Group by time of day and analyze accident counts
    time_of_day_counts = df['Sunrise_Sunset'].value_counts()
    print(time_of_day_counts)
else:
    print("Column 'Sunrise_Sunset' not found in the dataset.")

# Step 4: Visualizing Accident Hotspots

import folium
from folium.plugins import HeatMap

# Creating a map centered around the mean latitude and longitude of the accidents
accident_map = folium.Map(location=[df['Start_Lat'].mean(), df['Start_Lng'].mean()], zoom_start=5)

# Creating a HeatMap layer using the accident coordinates
heat_data = [[row['Start_Lat'], row['Start_Lng']] for index, row in df.iterrows()]
HeatMap(heat_data).add_to(accident_map)

# Displaying the map
accident_map.save('accident_heatmap.html')  # Save the map as an HTML file
```

```python
import pandas as pd

# Replace 'your_file_path' with the actual path to your CSV file
file_path = r'C:\Users\NHI652\Downloads\archive (2)\US_Accidents_March23.csv'
df = pd.read_csv(file_path)

# Display all column names in the dataset
print(df.columns)

# Step 3: Analyze Patterns

# Check if 'Weather_Condition' is present in the list of columns
if 'Weather_Condition' in df.columns:
    # Group by weather conditions and analyze accident counts
    weather_condition_counts = df['Weather_Condition'].value_counts()
    print(weather_condition_counts)
else:
    print("Column 'Weather_Condition' not found in the dataset.")

# Check if 'Sunrise_Sunset' is present in the list of columns
if 'Sunrise_Sunset' in df.columns:
    # Group by time of day and analyze accident counts
    time_of_day_counts = df['Sunrise_Sunset'].value_counts()
    print(time_of_day_counts)
else:
    print("Column 'Sunrise_Sunset' not found in the dataset.")

# Step 4: Visualize Accident Hotspots

try:
    import folium
    from folium.plugins import HeatMap

    # Create a map centered around the mean latitude and longitude of the accidents
    accident_map = folium.Map(location=[df['Start_Lat'].mean(), df['Start_Lng'].mean()], zoom_start=5)

    # Create a HeatMap layer using the accident coordinates
    heat_data = [[row['Start_Lat'], row['Start_Lng']] for index, row in df.iterrows()]
    HeatMap(heat_data).add_to(accident_map)

    # Display the map
    accident_map.save('accident_heatmap.html')  # Save the map as an HTML file

except ModuleNotFoundError:
    print("Folium module not found. Please install it using: pip install folium")
```
